```python
import numpy as np
import matplotlib.pyplot as plt

# Data
b = np.array([0.999862, 0.9999986, 0.9999999615, 0.99999999787])

# Compute Doppler shift
Z = np.sqrt((1+b)/(1-b)) - 1

# FLRW comoving distance approximation: R = c Z / ((1+Z) H0)
c = 3e5  # km/s
H0 = 70  # km/s/Mpc
R = c * Z / ((1 + Z) * H0)

# Plot
plt.figure()
plt.plot(Z, R, marker='o')
plt.xlabel("Redshift Z")
plt.ylabel("Comoving Distance R (Mpc)")
plt.title("Comoving Distance vs Redshift")
plt.tight_layout()
plt.savefig("distance_vs_z.pdf")
plt.show()
```
