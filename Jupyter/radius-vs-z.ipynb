```python
import numpy as np
import matplotlib.pyplot as plt

# Data: Sumerian units
units = ["Soss (60)", "Ner (600)", "Sar (3600)", "Shar (21600)"]
b_values = [0.999862, 0.9999986, 0.9999999615, 0.99999999787]

# Compute Doppler shift Z for each b
Z_values = [((1+b)/(1-b))**0.5 - 1 for b in b_values]

# Constants for radius computation
c = 3e5 # km/s
H0 = 70 # km/s/Mpc

# Compute radius for each Z
R_values = [c * Z / ((1 + Z) * H0) for Z in Z_values]

# Derived constant: age scale in Gyr
# 2 / (3 H0 sqrt(Omega_m)) converted to Gyr
# Using t_H  13.8 Gyr for normalization
age_scale_Gyr = 17.0  # consistent with CDM approximation
# Doppler-derived redshifts from the paper
Z = np.array([119, 1194, 7206, 30644], dtype=float)
age_Gyr = age_scale_Gyr * (1.0 + Z)**(-1.5)

# Plot
plt.figure()
plt.plot(units, R_values, marker='o', linestyle='-', color='orange')
plt.plot(units, R_values, marker='o', linestyle='-', color='orange')
plt.title('Estimated Universe Radius vs Sumerian Unit')
plt.xlabel('Sumerian Unit')
plt.ylabel('Radius R (Mpc)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("radius_vs_z.pdf")
plt.show()

plt.figure(figsize=(6,4))

plt.plot(units, age_Gyr, marker='o')

plt.yscale('log')
plt.xlabel("Sumerian Unit")
plt.ylabel("Age of Universe at Emission (Gyr)")
plt.title("Cosmic Age Implied by Sumerian Units")

plt.grid(True, which="both", ls="--", alpha=0.5)
plt.tight_layout()

plt.savefig("age_vs_units.pdf")
plt.show()
```
