```python
import numpy as np
import matplotlib.pyplot as plt

# Cosmology constants
H0 = 70 # km/s/Mpc
c = 3e5 # km/s

# Function to compute lookback distance and comoving distance (simplified flat universe, Omega=1)
def lookback_distance(z):
    return (c/H0) * z / (1+z) # approximate formula in Mpc

def comoving_distance(z):
    return (c/H0) * z # approximate formula in Mpc

# Redshift range for plot
z = np.linspace(0, 30, 300)
lookback = lookback_distance(z)
comoving = comoving_distance(z)

# Sumerian unit redshifts from previous calculations
unit_names = ["Soss (60)", "Ner (600)", "Sar (3600)", "Shar (21600)"]
Z_units = [119.41, 1194.24, 7206, 30644]

# Corresponding distances
lookback_units = [lookback_distance(Z) for Z in Z_units]
comoving_units = [comoving_distance(Z) for Z in Z_units]

# Plot Lookback vs Comoving Distance
plt.figure(figsize=(10,6))
plt.plot(z, lookback, label='Lookback Distance (Mpc)')
plt.plot(z, comoving, label='Comoving Distance (Mpc)')
plt.scatter(Z_units, lookback_units, color='red', label='Sumerian Units (Lookback)', marker='o')
plt.scatter(Z_units, comoving_units, color='purple', label='Sumerian Units (Comoving)', marker='x')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Redshift Z')
plt.ylabel('Distance (Mpc, log scale)')
plt.title('Lookback vs Comoving Distance with Sumerian Units')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
# plt.savefig("z_vs_units.pdf")
plt.show()
```
