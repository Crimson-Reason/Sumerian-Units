```python
# Sumerian Units Doppler Shift and Radius Plot

import matplotlib.pyplot as plt

# Data: Sumerian units
units = ["Soss (60)", "Ner (600)", "Sar (3600)", "Shar (21600)"]
b_values = [0.999862, 0.9999986, 0.9999999615, 0.99999999787]

# Compute Doppler shift Z for each b
Z_values = [((1+b)/(1-b))**0.5 - 1 for b in b_values]

# Constants for radius computation
c = 3e5 # km/s
H0 = 70 # km/s/Mpc

# Compute radius for each Z
R_values = [c * Z / ((1 + Z) * H0) for Z in Z_values]

# Plot b vs units
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(units, b_values, marker='o', linestyle='-', color='green')
plt.title('b vs Sumerian Unit')
plt.xlabel('Sumerian Unit')
plt.ylabel('b value (v/c)')
plt.ylim(0.9998, 1.0)
plt.grid(True, linestyle='--', alpha=0.6)


# Plot Doppler Shift Z
plt.subplot(1,3,2)
plt.plot(units, Z_values, marker='o', linestyle='-', color='teal')
plt.title('Doppler Shift Z vs Sumerian Unit')
plt.xlabel('Sumerian Unit')
plt.ylabel('Doppler Shift Z')
plt.grid(True, linestyle='--', alpha=0.6)


# Plot Radius R
plt.subplot(1,3,3)
plt.plot(units, R_values, marker='o', linestyle='-', color='orange')
plt.title('Estimated Universe Radius vs Sumerian Unit')
plt.xlabel('Sumerian Unit')
plt.ylabel('Radius R (Mpc)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("radius_z.pdf")

plt.tight_layout()
plt.show()
```
